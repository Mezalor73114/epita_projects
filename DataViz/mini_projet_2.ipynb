```python
import pandas as pd
import altair as alt
```

## __Mondial Statistics__

```python
data_total_views = data_youtube.groupby('category')['video views'].sum().reset_index()

alt.Chart(data_total_views).mark_bar().encode(
    x=alt.X('category', sort='-y'),
    y=alt.Y('video views', title='Video Views'),
    tooltip=[
        alt.Tooltip('category', title='Category'),
        alt.Tooltip('video views', title='Total Views', format=',')
    ]
).properties(
    title='Total Views by Category in 2023'
)
```

```python
category_counts = data_youtube['category'].value_counts().reset_index()
category_counts.columns = ['category', 'count']
category_counts = category_counts.sort_values(by='count', ascending=False)

chart = alt.Chart(category_counts).mark_arc().encode(
    theta=alt.Theta(field='count', type='quantitative'),
    color=alt.Color(field='category', type='nominal', sort=None),
    order=alt.Order('count:Q', sort='ascending'),
    tooltip=[
        alt.Tooltip('category', title='Category'),
        alt.Tooltip('count', title='Count', format=',')
    ]
).properties(
    title="Proportion of YouTube Categories by Channel Count",
    width=500,
    height=500
).configure_view(strokeWidth=0)

chart
```

```python
color_dict = {
    'Entertainment': '#E63946',
    'Music': '#A1FAEE',
    'People': '#A8DADC',
    'Games': '#457B9D',
    'Comedy': '#F4A261',
    'Education': '#2A9D8F',
    'Film': '#264653',
    'Howto': '#E9C46A',
    'News': '#F77F00',
    'Tech': '#2B2D42',
    'Sports': '#D90429',
    'Autos': '#8D99AE',
    'Animals': '#81B29A',
    'Nonprofit': '#3D405B'
}

type_counts = data_youtube["channel_type"].value_counts().reset_index()
type_counts.columns = ['channel_type', 'count']

chart = alt.Chart(type_counts).mark_bar().encode(
    x=alt.X('channel_type:N', 
            title='channel type', 
            sort=alt.EncodingSortField(
                field='count',
                order='descending'
            ),
            axis=alt.Axis(labelAngle=0)),
    y=alt.Y('count:Q', title='number of channels'),
    color=alt.Color('channel_type:N', scale=alt.Scale(domain=list(color_dict.keys()), range=list(color_dict.values())), legend=None)
).properties(
    title='Number of YouTube Channels by Type',
    width=650,
    height=300
)

chart
```

```python
subscribers_by_channel = data_youtube.groupby("channel_type")["subscribers"].mean().reset_index()
subscribers_by_channel.columns = ['channel_type', 'average_subscribers']

chart = alt.Chart(subscribers_by_channel).mark_bar().encode(
    x=alt.X('channel_type:N',
            title='channel type',
            sort=alt.EncodingSortField(
                field='average_subscribers',
                order='descending'
            ),
            axis=alt.Axis(labelAngle=0)),
    y=alt.Y('average_subscribers:Q',
            title='average number of subscribers'),
    color=alt.Color('channel_type:N',
                   scale=alt.Scale(
                       domain=list(color_dict.keys()),
                       range=list(color_dict.values())
                   ),
                   legend=None)
).properties(
    title='Average Number of Subscribers per Channel Type',
    width=650,
    height=300
)

chart
```

```python
data_year_sub = data_youtube.groupby('created_year')['subscribers'].mean().reset_index()
data_year_sub = data_year_sub.drop(data_year_sub[data_year_sub['created_year'] < 2004].index)


alt.Chart(data_year_sub).mark_line(point=alt.OverlayMarkDef(filled=False, fill="white")).encode(
    x=alt.X('created_year', title='Year', scale=alt.Scale(domain=[2004, 2023])),
    y=alt.Y('subscribers', title='Mean Subscribers'),
    tooltip=[
        alt.Tooltip('created_year', title='Year'),
        alt.Tooltip('subscribers', title='Mean Subscribers')
    ]
).properties(
    title='Average Number of Subscribers by Year',
    height=500,
    width=700
)
```

```python
youtuber_count_by_created_year = data_youtube.groupby('created_year')['Youtuber'].count().reset_index()

chart = alt.Chart(youtuber_count_by_created_year).mark_bar().encode(
    x=alt.X('created_year:O', axis=alt.Axis(labelAngle=45)),
    y=alt.Y('Youtuber:Q', title='Youtuber Count'),
    color=alt.value('steelblue')
).properties(
    title='Number of YouTube Channels Created per Year',
    width=600,
    height=400
).configure_axis(
    labelFontSize=10
)

chart
```

```python
data_drop_sub_views_cat = data_youtube.dropna(subset=['subscribers', 'video views', 'category'])

alt.Chart(data_drop_sub_views_cat).mark_rect().encode(
    x=alt.X('subscribers', title='Subscribers', bin=alt.Bin(maxbins=100)),
    y=alt.Y('video views', title='Video Views', bin=alt.Bin(maxbins=100)),
    color=alt.Color('category', title='Category'),
    tooltip=[
        alt.Tooltip('Youtuber', title='Youtuber'),
        alt.Tooltip('subscribers', title='Subscribers'),
        alt.Tooltip('video views', title='Video Views'),
        alt.Tooltip('category', title='Category')
    ]
).properties(
    title='Subscribers vs Video Views by Category',
    width=700,
    height=500
).configure_axis(
    grid=True
)
```

## __By country statistics__

```python
data_count_youtuber = data_youtube.groupby('Country')['Youtuber'].count().reset_index()
data_country_sub = data_youtube.groupby('Country')['subscribers'].sum().reset_index()
data_country_views = data_youtube.groupby('Country')['video views'].sum().reset_index()


chart1 = alt.Chart(data_count_youtuber).mark_bar().encode(
    x=alt.X('Country', sort='-y'),
    y=alt.Y('Youtuber', title='Number of Youtubers'),
    tooltip=[
        alt.Tooltip('Country', title='Country'),
        alt.Tooltip('Youtuber', title='Number of Youtubers', format=',')
    ]
).properties(
    title='Number of Youtubers by Country'
)

chart2 = alt.Chart(data_country_sub).mark_bar().encode(
    x=alt.X('Country', sort='-y'),
    y=alt.Y('subscribers', title='Total Subscribers'),
    tooltip=[
        alt.Tooltip('Country', title='Country'),
        alt.Tooltip('subscribers', title='Total Subscribers', format=',')
    ]
).properties(
    title='Total Subscribers by Country'
)

chart3 = alt.Chart(data_country_views).mark_bar().encode(
    x=alt.X('Country', sort='-y'),
    y=alt.Y('video views', title='Total Video Views'),
    tooltip=[
        alt.Tooltip('Country', title='Country'),
        alt.Tooltip('video views', title='Total Video Views', format=',')
    ]
).properties(
    title='Total Video Views by Country'
)

chart1 & chart2 & chart3
```

```python
from vega_datasets import data

data_youtube_grouped = data_youtube.groupby(['Country', 'channel_type'])['subscribers'].sum().reset_index()
data_youtube_dominant = data_youtube_grouped.loc[data_youtube_grouped.groupby('Country')['subscribers'].idxmax()]

data_youtube_pop = data_youtube.groupby('Country').agg({
    'Population': 'first'
}).reset_index()
data_youtube_dominant = data_youtube_dominant.merge(data_youtube_pop, on='Country')

world = alt.topo_feature(data.world_110m.url, 'countries')

# Fond de carte
base = alt.Chart(world).mark_geoshape(
    fill='lightgray',
    stroke='darkgrey',
    strokeWidth=0.5
).project(
    type='equirectangular'
).properties(
    width=800,
    height=400
)

countries = alt.Chart(world).mark_geoshape().encode(
    # couleur grise si country est null
    color=alt.condition(
        'isDefined(datum.Country) && datum.Country != null',
        alt.Color('channel_type:N', legend=alt.Legend(title='Channel Type'),
                  scale=alt.Scale(domain=[v for v in data_youtube_dominant['channel_type'].unique() if v is not None])),
        alt.value('lightgray')
    ),
    tooltip=['Country:N', 'channel_type:N', 'subscribers:Q', 'Population:Q']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(
        data=country_codes,
        key='id',
        fields=['name']
    )
).transform_lookup(
    lookup='name',
    from_=alt.LookupData(
        data=data_youtube_dominant,
        key='Country',
        fields=['Country', 'channel_type', 'subscribers', 'Population']
    )
).project(
    type='equirectangular'
).properties(
    width=800,
    height=400,
    title="Dominant YouTube Channel Types by Country"
)

base + countries
```

## __Socio-economical statistics__

```python
data_sum_sub_pop = data_youtube.dropna(subset=['subscribers', 'Population', 'Country'])

data_sum_sub_pop = data_sum_sub_pop.groupby(['Country', 'Population'])['subscribers'].mean().reset_index()

alt.Chart(data_sum_sub_pop).mark_circle(size=50).encode(
    x=alt.X('Population', title='Population'),
    y=alt.Y('subscribers', title='Mean Subscribers'),
    color=alt.Color('Country', title='Country'),
    tooltip=[
        alt.Tooltip('Country', title='Country'),
        alt.Tooltip('Population', title='Population'),
        alt.Tooltip('subscribers', title='Mean Subscribers')
    ]
).properties(
    title='Mean Subscribers vs Population',
    width=700,
    height=500
)
```

```python
data_mean_views_pop = data_youtube.dropna(subset=['video views', 'Population', 'Country'])

data_mean_views_pop = data_mean_views_pop.groupby(['Country', 'Population'])['video views'].mean().reset_index()

alt.Chart(data_mean_views_pop).mark_circle(size=50).encode(
    x=alt.X('Population', title='Population'),
    y=alt.Y('video views', title='Mean Video Views'),
    color=alt.Color('Country', title='Country'),
    tooltip=[
        alt.Tooltip('Country', title='Country'),
        alt.Tooltip('Population', title='Population'),
        alt.Tooltip('video views', title='Mean Video Views')
    ]
).properties(
    title='Mean Video Views vs Population',
    width=700,
    height=500
)
```

```python
data_unemployement = data_youtube.dropna(subset=['Country', 'Unemployment rate'])
data_unemployement = data_unemployement.groupby('Country').agg({'video views': "sum", "Unemployment rate": "mean"}).reset_index()

data_unemployement

alt.Chart(data_unemployement).mark_circle(size=50).encode(
    x = alt.X('Unemployment rate'),
    y = alt.Y('video views', title='Video views'),
    color='Country',
    tooltip=['Country', 'Unemployment rate', 'video views']
).properties(
    title='Video views by unemployment rate of the country',
    width=500,
    height=500
)
```

```python
def most_frequent_category(series):
    return series.mode()[0] if not series.mode().empty else None

data_unemployement_cat = data_youtube.dropna(subset=['Country', 'Unemployment rate'])
data_unemployement_cat = data_unemployement_cat.groupby('Country').agg({'Unemployment rate': 'mean', 'category': most_frequent_category}).reset_index().sort_values("Unemployment rate", ascending=False)
data_unemployement_cat

alt.Chart(data_unemployement_cat[:20]).mark_bar().encode(
    x=alt.X('Country', sort=alt.EncodingSortField(field='Unemployment rate', order='ascending')),
    y=alt.Y('Unemployment rate'),
    color=alt.Color('category', title='Category'),
    tooltip=['Country', 'Unemployment rate', 'category']
).properties(
    title='Most popular category for the country with the hightest unemployment rate',
    width=500,
    height=500
)
```

```python
data_urbasided = data_youtube.dropna(subset=['Country', 'Urban_population', 'category'])
data_urbasided = data_urbasided.groupby('Country').agg({'Urban_population': 'mean', 'category': most_frequent_category}).reset_index().sort_values("Urban_population", ascending=False)

alt.Chart(data_urbasided[:20]).mark_bar().encode(
    x=alt.X('Country', sort=alt.EncodingSortField(field='Urban_population', order='ascending')),
    y=alt.Y('Urban_population'),
    color=alt.Color('category'),
    tooltip=['Country', 'Urban_population', 'category']
).properties(
    title='Most popular category by the number of urbanased population of the country',
    width=1000,
    height=500
)
```
